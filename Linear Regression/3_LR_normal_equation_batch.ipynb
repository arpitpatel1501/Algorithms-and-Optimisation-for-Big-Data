This code is generalized for both Simple and Multiple Linear Regression accrding to the input data X


Importing Libraries

```
import numpy as np                              #import for numerical opoeration
from sklearn import datasets, metrics           #import for boston data and matrix operation such as multiplication and other
from numpy.linalg import inv, pinv, LinAlgError #import for finding inverse and psuedo inverse of matrix form numpy linear algebra
from sklearn.preprocessing import StandardScaler#import for z score normalization
```

Load the Boston dataset from sklearn library

```
X,y = datasets.load_boston(return_X_y=True)
X.shape
```

Diving dataset into training and testing (hold out) set

```
TEST_SPLIT = 0.8                    # split data into 80%, 20% into training and testind respectively

X_train_temp = X[0:int(len(X)*TEST_SPLIT),:]                                           #Take first 400 example(rows) and all the feature(column) for training 
X_train = np.zeros((X_train_temp.shape[0],X_train_temp.shape[1]+1)) #.shape function gives an dimenasion so we assign the dimension of the X_train_temp 
                                                                    #and add one more feature as dummy feature so add one into the column part
X_train[:,0] = np.ones((X_train.shape[0]))                          # Add 1 as dummy feature in all examples
X_train[:,1:] = X_train_temp
print("Type of X_train : ",type(X_train))
print("Size of the X_train : ",X_train.shape)

y_train = y[0:int(len(X)*TEST_SPLIT)]

X_test_temp = X[int(len(X)*TEST_SPLIT):,:]                                           #Take remaining 106 example(rows) and all the feature(column) for training 
X_test = np.zeros((X_test_temp.shape[0],X_test_temp.shape[1]+1)) #.shape function gives an dimenasion so we assign the dimension of the X_test_temp 
                                                                    #and add one more feature as dummy feature so add one into the column part
X_test[:,0] = np.ones((X_test.shape[0]))                          # Add 1 as dummy feature in all examples
X_test[:,1:] = X_test_temp
print("Type of X_test : ",type(X_test))
print("Size of the X_test : ",X_test.shape)


y_test = y[int(len(X)*TEST_SPLIT):]
```

Normalize the data : Z score normalization

>$v' =\frac{(v - A\bar)}{\sigma_A}$  
where  
>$\sigma_A^2 = \frac{1}{N} \sum_{i=1}^{n}(x_i- x^- )^2$

```
scaler=StandardScaler()
scaler.fit(X_train[:,1:])
X_train[:,1:]=scaler.transform(X_train[:,1:])
X_test[:,1:]=scaler.transform(X_test[:,1:])
```

Parameter theta of Regression model:
>$\theta = (X^TX)^{-1}X^Ty$

```
def LR_simple_multiple(X_train):
  theta=np.zeros(X_train.shape[1])

  try:
    XTXi=inv(np.dot(X_train.T,X_train))
  except LinAlgError:
      XTXi=pinv(np.dot(X_train.T,X_train))

  XTy=np.dot(X_train.T,y_train)    

  theta=np.dot(XTXi,XTy)


  print("Theta:",theta)
  print("Theta Shape:", theta.shape)

  return theta

```

Prediction :
 >$y = X \theta$

```
theta = LR_simple_multiple(X_train)
predictions=np.dot(X_test, theta)
```

Mean Square Error and Mean Absolute Error

```
print("MAE:", metrics.mean_absolute_error(y_true=y_test,y_pred=predictions))
print("MSE:", metrics.mean_squared_error(y_true=y_test,y_pred=predictions))
```
